```python
df1_adam = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/adam_weights_0.csv')
df2_adam = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/adam_weights_01.csv')
df3_adam = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/adam_weights_-01.csv')
df4_adam = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/adam_weights_001.csv')
df5_adam = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/adam_weights_-001.csv')

df1_gd = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/gd_weights_0.csv')
df2_gd = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/gd_weights_01.csv')
df3_gd = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/gd_weights_-01.csv')
df4_gd = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/gd_weights_001.csv')
df5_gd = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/gd_weights_-001.csv')

df1_sgd = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/sgd_weights_0.csv')
df2_sgd = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/sgd_weights_01.csv')
df3_sgd = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/sgd_weights_-01.csv')
df4_sgd = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/sgd_weights_001.csv')
df5_sgd = pd.read_csv('C:/Users/kaany/Desktop/Uni Shit/YTU/3rd Semester/BLM2642/imageClassification/imageClassification/weights/sgd_weights_-001.csv')



df_all_adam = pd.concat([df1_adam, df2_adam, df3_adam, df4_adam, df5_adam], axis=0)
df_all_gd = pd.concat([df1_gd, df2_gd, df3_gd, df4_gd, df5_gd], axis=0)
df_all_sgd = pd.concat([df1_sgd, df2_sgd, df3_sgd, df4_sgd, df5_sgd], axis=0)
# display(df_all)

y = []
for i in [0, 0.1, -0.1, 0.01, -0.01]:
    for j in range(500):
        y.append(i)
```

```python
model = TSNE(n_components = 2, random_state = 0)
tsne_data_adam = model.fit_transform(df_all_adam)
tsne_data_adam = np.vstack((tsne_data_adam.T, y)).T
tsne_df_adam = pd.DataFrame(data = tsne_data_adam, columns =('t-SNE Feature 1', 't-SNE Feature 2', 'Starting weights'))

tsne_data_gd = model.fit_transform(df_all_gd)
tsne_data_gd = np.vstack((tsne_data_gd.T, y)).T
tsne_df_gd = pd.DataFrame(data = tsne_data_gd, columns =('t-SNE Feature 1', 't-SNE Feature 2', 'Starting weights'))

tsne_data_sgd = model.fit_transform(df_all_sgd)
tsne_data_sgd = np.vstack((tsne_data_sgd.T, y)).T
tsne_df_sgd = pd.DataFrame(data = tsne_data_sgd, columns =('t-SNE Feature 1', 't-SNE Feature 2', 'Starting weights'))
 
# Plotting the result of tsne
sns.scatterplot(data=tsne_df_adam, x='t-SNE Feature 1', y='t-SNE Feature 2', hue='Starting weights', palette='bright').set_title('t-SNE Visualization')

sns.scatterplot(data=tsne_df_gd, x='t-SNE Feature 1', y='t-SNE Feature 2', hue='Starting weights', palette='hls').set_title('t-SNE Visualization ') 

sns.scatterplot(data=tsne_df_sgd, x='t-SNE Feature 1', y='t-SNE Feature 2', hue='Starting weights').set_title('t-SNE Visualization ') 

plt.legend(title='Starting Weights', labels=['ADAM 0', 'ADAM 0.1', 'ADAM -0.1', 'ADAM 0.01', 'ADAM -0.01',
                                             'GD 0', 'GD 0.1', 'GD -0.1', 'GD 0.01', 'GD -0.01',
                                             'SGD 0', 'SGD 0.1', 'SGD -0.1', 'SGD 0.01', 'SGD -0.01'])
plt.show()
```
